```python
# Dependencies
import tweepy
import twitter_service as ts
import analysis_service as at
from collections import defaultdict
from config import consumer_key, consumer_secret, access_token, access_token_secret
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Twitter API Keys
consumer_key = consumer_key
consumer_secret = consumer_secret
access_token = access_token
access_token_secret = access_token_secret
```

```python
handles = ['@CNN', '@BBCWorld', '@FoxNews', '@CBSNews', '@nytimes']

    # Variables for holding sentiments
sentiments = defaultdict(list)

    # Variable for max_id
oldest_tweet = None
    
api = ts.setup_connection(consumer_key, consumer_secret, access_token, access_token_secret)
    
for handle in handles:
    public_tweets = ts.grab_tweets(api, handle, 100)
    for index, tweet in enumerate(public_tweets):
        # Run Vader Analysis on each tweet
        results = at.analyze_tweets(tweet)
        sentiments["Handle"].append(tweet["user"]["screen_name"])
        sentiments["Date"].append(tweet["created_at"])
        sentiments["Compound"].append(results["compound"])
        sentiments["Positive"].append(results["pos"])
        sentiments["Neutral"].append(results["neu"])
        sentiments["Negative"].append(results["neg"])
        sentiments["Tweets Ago"].append(index + 1)
        sentiments["Tweet Text"].append(tweet["text"])
    
print("Finished!")
    
```

```python
def create_dataframe(data):
    return pd.DataFrame(data)
        
tweet_sentiment_df = create_dataframe(sentiments)
```

```python
cnn_compound = tweet_sentiment_df.loc[tweet_sentiment_df['Handle'] == 'CNN']

bbc_compound = tweet_sentiment_df.loc[tweet_sentiment_df['Handle'] == 'BBCWorld']

fox_compound = tweet_sentiment_df.loc[tweet_sentiment_df['Handle'] == 'FoxNews']

cbs_compound = tweet_sentiment_df.loc[tweet_sentiment_df['Handle'] == 'CBSNews']

nyt_compound = tweet_sentiment_df.loc[tweet_sentiment_df['Handle'] == 'nytimes']
```

```python
#convert createdate of first tweet to mm/dd/yyyy format for display on graphs
record_date = datetime.strptime(tweet_sentiment_df['Date'][0],'%a %b %d %H:%M:%S +0000 %Y').strftime('%m/%d/%Y')

print(record_date)
```
